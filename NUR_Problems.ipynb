```python
import pandas as pd

df = pd.read_csv('../Downloads/National_Universities_Rankings.csv')
pd.set_option('display.max_columns',8)
pd.set_option('display.max_rows',231)
df
```

```python
# 1.Find the universities in which undergraduate students were enrolled.
f = df['Undergrad Enrollment'].isnull()
df[~f]['Name']
```

```python
# 2.Find the universities in which In-state students were enrolled.
f = df['In-state'].isnull()
df[~f]['Name']
```

```python
# 3.Find the universities in which in-state students were not enrolled
f = df['In-state'].isnull()
df[f]['Name'].describe()
```

```python
# 4.How many universities have both out-of state and in-state students?
f1 = ~(df['In-state'].isnull()) & ~(df['Tuition and fees'].isnull()) 
df[f1]
```

```python
# 5.List the the cities in which universities were  located.
df[['city','state']] = df.Location.str.split(',',expand=True)
df[['Name','city']]
```
